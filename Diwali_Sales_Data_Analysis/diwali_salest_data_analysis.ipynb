```python
# Importing relevant python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df - represents a dataframe. A DataFrame is a data structure that organizes data into a 2-dimensional table of 
# rows and columns, much like a spreadsheet. We are passing the path to our data source (local path in this case)
df = pd.read_csv('data_sources/diwali_sales_data.csv', encoding='unicode_escape')
```

```python
# checks the number of rows and columns in the dataframe.
df.shape
```

```python
# The head() function returns five rows by default. If we have to put any fixed number, we can put it like this:
# head(10) - this will return 10 rows.
df.head()
```

```python
# Getting first ten rows
df.head(10)
```

```python
# Provides information about the columns, count of non-null and null rows and data types, column headers, etc.
df.info()
```

```python
# As we can see that last two columns are not adding any insights to this dataframe, hence, we will delete it by using 
# drop function.
# inplace=True is used to save the changes made.
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.shape # here, we can see that number of columns remained 13 and earlier it was 15.
```

```python
# We also see that 'Amount' column is only having 11239 rows and others are having 11251. 
# To remove null values, first we need to find out how many null values are there. 
# We will use isnull() and sum() function to calculate the same. 
df.info() 
```

```python
df.isnull().sum() 
```

```python
# Amount column is having 12 null values, we will have to delete it. We will use dropna() function to delete the null value.
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# We can see that the number of rows and columns have been updated now. 
```

```python
# This function is used to check the name of the columns.
df.columns
```

```python
# Now we want to analyze the data on the basis of Gender, out of 11239 rows of customes, which Gender has done more shopping.
# We will use countplot from seaborn library.
ax=sns.countplot(x='Gender', data=df)
```

```python
# We can see number of females are more than the numbers of males but count is not being displayed. 
# To displaye the count we will use below loop.
ax=sns.countplot(x='Gender', data=df)

for bars in ax.containers: ax.bar_label(bars)
```

```python
# Similarly, we can perforn this analysis statewise or zone wise as mentioned below.
ax=sns.countplot(x='Zone', data=df)
for bars in ax.containers: ax.bar_label(bars)
```

```python
ax=sns.countplot(x='State', data=df)

for bars in ax.containers: ax.bar_label(bars)
```

```python
# We are not able to see the x-axis clearly, so we will set the width and lenght of the bars using below program:
ax=sns.countplot(x='State', data=df, hue='Gender')

sns.set(rc={'figure.figsize':(30,5)})

# The attribute hue='Gender', will further devide one particular bar of particular state in to male and female.
for bars in ax.containers: ax.bar_label(bars)
```

```python
ax=sns.countplot(x='State', data=df)

sns.set(rc={'figure.figsize':(30,5)})

for bars in ax.containers: ax.bar_label(bars)
```

```python
# Group by function to analysis total purchasing done on the basis of Gender. 
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
x=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount', data=x)
```

```python
# As we have set the width and length of bars in the last promgram, it has taken the same as default value. 
# To change it again, we will use set function.
x=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(3,4)})
sns.barplot(x='Gender',y='Amount', data=x)
```

```python
# We want to analyse total purchase state wise for tot 10 states, for this we will follow the program as mentioned below.
x=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
                                                                                                    
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount', data=x)
```
